```python
from msilib import sequence
from turtle import color
import numpy as np
import statsmodels.api as sm
import matplotlib as plt
from matplotlib import pyplot
from scipy.stats import norm
import random

# Primeiro implementaremos o Q-Q plot e depois utilizaremos o pacote statsmodels para plottar
random.seed(343392)
measurements = np.random.normal(loc = 20, scale = 5, size=100)

def abline(slope, intercept):
    axes = plt.pyplot.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.pyplot.plot(x_vals, y_vals, '--',color = "#ff0000")

def qq_plot(data):
    n = np.size(data)
    x_ordered = np.sort(data)
    j = np.linspace(1,n,n,endpoint = True)
    pj = (j-0.5)/ n
    qj = norm.ppf(pj)
    Qa = np.quantile(data,[0.25,0.75])
    Qt = norm.ppf([0.25,0.75])
    slope = np.diff(Qa)/np.diff(Qt)
    inter = Qa[0] - slope*Qt[0]
    plt.pyplot.scatter(qj,x_ordered,s = 10, color = "#000000")
    abline(slope,inter)
    return 
    
print (qq_plot(measurements))
```
